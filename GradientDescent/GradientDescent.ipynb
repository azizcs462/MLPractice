```python
class BGradientDescent:
    def __init__(self,epochs=100,learning_rate=0.01):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.coef_ = None
        self.intercept_ = None

    def fit(self,X_train,y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        print(self.coef_.shape)
        for i in range(self.epochs):
            y_hat = np.dot(X_train,self.coef_)+self.intercept_
            # print(y_hat.shape)
            intercept_der = -2 * np.mean(y_train-y_hat)
            self.intercept_ = self.intercept_ - (self.learning_rate*intercept_der)
            
            coef_der = -2*np.dot((y_train-y_hat),X_train)/X_train.shape[0] #Vectorization
            self.coef_ = self.coef_ - (self.learning_rate*coef_der)
        print(self.coef_,self.intercept_)

    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_


            
```

```python
BGD = BGradientDescent(epochs=1005,learning_rate=0.45)
```

```python
BGD.fit(X_train,y_train)
```

```python
y_pred = BGD.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
class SGradientDescent:
    def __init__(self,epochs=100,learning_rate=0.01):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.coef_ = None
        self.intercept_ = None

    def fit(self,X_train,y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        print(self.coef_.shape)
        for i in range(self.epochs):
            for j in range(X_train.shape[0]):
                idx = np.random.randint(0,X_train.shape[0])
                y_hat = np.dot(X_train[idx],self.coef_)+self.intercept_
                intercept_der = -2 * (y_train[idx]-y_hat)
                coef_der = -2*np.dot((y_train[idx]-y_hat),X_train[idx]) #Vectorization

                self.intercept_ = self.intercept_ - (self.learning_rate*intercept_der)
                self.coef_ = self.coef_ - (self.learning_rate*coef_der)
        print(self.coef_,self.intercept_)
       
            

    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_


            
```

```python
SGD = SGradientDescent(epochs=55,learning_rate=0.03)
```

```python
SGD.fit(X_train,y_train)
```

```python
y_pred = SGD.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

# <font style color = 'red'>MiniBatch GradientDescent</font>

```python
import random
class MBGradientDescent:
    def __init__(self,batch_size,epochs=100,learning_rate=0.01):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.coef_ = None
        self.intercept_ = None
        self.batch_size = batch_size

    def fit(self,X_train,y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        # print(self.coef_.shape,self.batch_size)
        for i in range(self.epochs):
            for j in range(int(X_train.shape[0]/self.batch_size)):
                idx = random.sample(range(X_train.shape[0]),self.batch_size)
                # print(idx)
                y_hat = np.dot(X_train[idx],self.coef_)+self.intercept_
                intercept_der = -2 * np.mean(y_train[idx]-y_hat)
                coef_der = -2*np.dot((y_train[idx]-y_hat),X_train[idx]) #Vectorization

                self.intercept_ = self.intercept_ - (self.learning_rate*intercept_der)
                self.coef_ = self.coef_ - (self.learning_rate*coef_der)
        print(self.intercept_)
       
            

    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_


            
```

```python
mbgd = MBGradientDescent(epochs=50,learning_rate=0.01,batch_size=int(X_train.shape[0]/10))
mbgd.fit(X_train,y_train)
```

```python
y_pred = mbgd.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
print(nlr.coef_,nlr.intercept_)
```

```python
rlr = Ridge(alpha=10)
rlr.fit(X,y)
print(rlr.coef_,rlr.intercept_)
```

```python
rlr1 = Ridge(alpha=100)
rlr1.fit(X,y)
print(rlr.coef_,rlr.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,nlr.predict(X),color='red',label = 'alpha=0')
plt.plot(X,rlr.predict(X),color='green',label = 'alpha=10')
plt.plot(X,rlr1.predict(X),color='blue',label = 'alpha=100')


plt.legend()
plt.show()
```

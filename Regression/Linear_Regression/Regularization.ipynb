```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
#plot data
plt.scatter(X,y)
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=.20)
```

```python
nlr = LinearRegression()
nlr.fit(X_train,y_train)
```

```python
print(nlr.coef_,nlr.intercept_)
```

```python
rlr = Ridge(alpha=10)
rlr.fit(X_train,y_train)
print(rlr.coef_,rlr.intercept_)
```

```python
rlr1 = Ridge(alpha=1000)
rlr1.fit(X_train,y_train)
print(rlr.coef_,rlr.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,nlr.predict(X),color='red',label = 'alpha=0')
plt.plot(X,rlr.predict(X),color='green',label = 'alpha=10')
plt.plot(X,rlr1.predict(X),color='blue',label = 'alpha=100')


plt.legend()
plt.show()
```

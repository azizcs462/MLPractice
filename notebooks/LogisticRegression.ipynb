```python
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,
                          random_state=41,class_sep=30,hypercube=False)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def step(n):
    return 1 if n>0 else 0
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.01

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]    
        
    
```

```python
intercept_,coeff_ = perceptron(X,y)
intercept_,coeff_
```

```python
m = -(coeff_[0]/coeff_[1])
b = -(intercept_/coeff_[1])
m,b
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input+b
```

-3 = starting point

3 = ending point

100 = number of values

So, it creates a NumPy array with 100 points between -3 and 3, spaced evenly. This is often used for:

Plotting smooth curves

Feeding input to functions for visualization (e.g., sigmoid, Gaussian, ML model decision boundary)



```python
plt.Figure(figsize=(10,6))
plt.plot(x_input,y_input,color = 'red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=50)
plt.ylim(-3,2)
```

X

X[:,0]  all values from the first column of your input X (i.e., x-axis)

X[:,1]  all values from the second column of X (i.e., y-axis)

c=y  color each point based on its label or class from y

cmap='winter'  use the 'winter' colormap (blue-green gradient)

s=100  size of each point is 100 (big enough to be visually clear)

```python
def sigmoid(n):
    return 1/(1+np.exp(-n))
```

```python
def perceptron1(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X[1].shape)
    lr = 0.01
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = sigmoid(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]    
```

```python
intercept_,coeff_ = perceptron1(X,y)
m = -(coeff_[0]/coeff_[1])
b = -(intercept_/coeff_[1])
```

```python
x_input2 = np.linspace(-3,3,100)
y_input2 = m*x_input2+b
```

```python
plt.Figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.plot(x_input2,y_input2,color='grey',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

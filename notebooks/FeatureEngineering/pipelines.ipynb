```python
#Missing Value Imputer Transformer
tf1 = ColumnTransformer(transformers=[('age_imputer',SimpleImputer(),[2]),
                        ('embarked_imputer',SimpleImputer(strategy='most_frequent'),[6])],remainder='passthrough')
```

```python
#Category to numerical OneHot Encoder
tf2 = ColumnTransformer([('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])],
                         remainder='passthrough')
```

```python
X_train
```

```python
#Feature Scaling minMax Transformer
tf3 = ColumnTransformer([('scale',MinMaxScaler(),slice(0,10))])
```

```python
#Feature Selection
tf4 = SelectKBest(score_func=chi2,k=5)
```

```python
#Model 
tf5 = DecisionTreeClassifier()
```

```python
pipe = Pipeline([('tf1',tf1),
                 ('tf2',tf2),
                 ('tf3',tf3),
                 ('tf4',tf4),
                 ('tf5',tf5)
                 ])
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(tf1,tf2,tf3,tf5)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

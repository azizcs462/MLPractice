```python
region_df = []

# Loop through each region
for region_var in all_trips['Pick_Location_Sub_Service_Region'].unique():
    print(f"Processing region: {region_var}")
    
    # Filter rows for this region
    Region_df_var = all_trips[all_trips['Pick_Location_Sub_Service_Region'] == region_var].copy()
    
    # Sort by Trip_Date to maintain time order
    Region_df_var = Region_df_var.sort_values('Trip_Date')
    
    # Calculate rolling sum of previous 4 entries (excluding current)
    Region_df_var['Last1Hour_trips'] = (
        Region_df_var['All_Rides_created_trips']
        .shift(1)  # shift to exclude the current row
        .rolling(window=4, min_periods=1)
        .sum()
    )
    
    # Append to list
    region_df.append(Region_df_var)

# Concatenate all regional data back into one DataFrame
all_trips = pd.concat(region_df, axis=0).sort_index()
```

```python
def generate_exponential_weights(n, alpha=0.01):
    """
    Generates exponentially increasing weights.
    
    Parameters:
    n (int): Number of data points.
    alpha (float): Decay rate. Higher values increase weights faster.
    
    Returns:
    numpy.ndarray: Array of weights.
    """
    return np.exp(alpha * np.arange(n))


# COMMAND ----------

from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# X_train is defined in your code
categorical_cols = ['Pick_Location_Sub_Service_Region']  
```

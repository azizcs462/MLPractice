```python
from matplotlib import pyplot as plt
plt.style.use('ggplot')
import numpy as np
import random
```

```python
data_X = [i*np.random.randint(0,2) for i in range(10,20)]
data_Y = [i*np.random.randint(20,30) for i in range(10,20)]
```

```python
plt.bar(data_X,data_Y,label = 'bar plot',color='b',linestyle='--')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('bar plot')
plt.legend()
plt.show()
```

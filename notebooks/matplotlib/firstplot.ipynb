```python
dev_x = [i for i in range(20,50)]
dev_y = [i+20 for i in range(50000,50030)]
dev_z = [i+30 for i in range(50000,50030)]
```

```python
# fmt_string [marker][line][color]
#can use hex code for colour also
plt.plot(dev_x,dev_y,color='k',linestyle = '--',marker = '.',label = 'all',linewidth=2)
plt.plot(dev_x,dev_z,'o:b',label = 'python')
plt.title('Salary wrt age')
plt.xlabel('Ages')
plt.ylabel('Salary')
# plt.legend(['All Devs','Python']) #we can give legend like this also
plt.legend()
plt.grid(True)

plt.tight_layout() #padding
# plt.savefig('plot1.png') #save png
plt.show()
```
